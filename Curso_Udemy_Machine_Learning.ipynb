```python
#reemplazando los nan con el promedio de cada columna

# Creando el DataFrame con los datos proporcionados
diccionario = {'A': [4, 5, np.nan],
               'B': [6, 1, 5],
               'C': [np.nan, 4, np.nan]}
dataframe = pd.DataFrame(diccionario)

# Calculando el promedio de cada columna
promedios = dataframe.mean()

# Reemplazando los NaN con el promedio de cada columna
dataframe.fillna(promedios, inplace=True)

print(dataframe)
```

```python
#combinando DF
import pandas as pd
df1 = {'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]}
concatenacion1 = pd.DataFrame(df1)
concatenacion1
```

```python
df2 = {'A':[11,12,13],'B':[14,15,16],'C':[17,18,19]}
concatenacion2 = pd.DataFrame(df2)
concatenacion2
```

```python
#para concatenar usamos el metodo concat()
pd.concat([concatenacion1,concatenacion2])
```

```python
#los concatenaremos uno al lado del otro
pd.concat([concatenacion1,concatenacion2], axis = 1)
```

```python
#uso de merge
import numpy as np
import pandas as pd
```

```python
merge1 = {'A':[1,2,3],
          'B':[4,5,6],
          'clave':['c1','c2','c3']}
union1 = pd.DataFrame(merge1)
union1
```

```python
merge2 = {'C':[11,12,13],
          'D':[14,15,16],
          'clave':['c1','c2','c3']}
union2 = pd.DataFrame(merge2)
union2
```

```python
#vamos a usar merge o union.
pd.merge(union1,union2, on='clave')
```

```python
#uso de join

lista1 = {'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]}
dataframe1 = pd.DataFrame(lista1, index = ['i1','i2','i3'])
dataframe1
```

```python
lista2 = {'D':[10,11,12],
          'E': [13,14,15],
          'F':[16,17,18]}
dataframe2 = pd.DataFrame(lista2, index = ['i1','i2','i3'])
dataframe2
```

```python
dataframe1.join(dataframe2)
```
